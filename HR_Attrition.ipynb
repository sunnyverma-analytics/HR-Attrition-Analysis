**Attrition Pie Chart**

```python
import matplotlib.pyplot as plt

attrition = df['Attrition'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    attrition,
    labels=attrition.index,
    autopct='%1.1f%%',
    colors=['red','skyblue']
)

plt.title("Employee Attrition Distribution")
plt.show()
```

**Average Salary by Department**

```python
dept_salary = df.groupby('Department')['MonthlyIncome'].mean()

plt.figure(figsize=(10,6))

plt.bar(
    dept_salary.index,
    dept_salary.values,
    color=['red','green','blue']
)

plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')

plt.show()
```

**Job Role Employee Count**

```python
job_counts = df['JobRole'].value_counts()

plt.figure(figsize=(12,6))

plt.barh(
    job_counts.index,
    job_counts.values,
    color='orange'
)

plt.title('Employees by Job Role')

plt.show()
```

**Age vs Income Scatter Plot**

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df['Age'],
    df['MonthlyIncome'],
    color='green'
)

plt.title('Age vs Monthly Income')
plt.xlabel('Age')
plt.ylabel('Income')

plt.show()
```

**Seaborn Visualizations**

```python
import seaborn as sns
```

**Overtime vs Attrition**

```python
plt.figure(figsize=(8,6))

sns.countplot(
    x='OverTime',
    hue='Attrition',
    data=df,
    palette='Paired'
)

plt.title('Overtime Impact on Attrition')

plt.show()
```

**Job Satisfaction vs Attrition**

```python
plt.figure(figsize=(10,6))

sns.countplot(
    x='JobSatisfaction',
    hue='Attrition',
    data=df,
    palette='rocket'
)

plt.title('Job Satisfaction vs Attrition')

plt.show()
```

**Income vs Attrition**

```python
plt.figure(figsize=(10,6))

sns.boxplot(
    x='Attrition',
    y='MonthlyIncome',
    data=df,
    palette='Set3'
)

plt.title('Income by Attrition Status')

plt.show()
```

**Correlation Heatmap**

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')

plt.show()
```

**Attrition By Department**

```python
dept_attrition = pd.crosstab(
    df['Department'],
    df['Attrition'],
    normalize='index'
)*100

print(dept_attrition)
```

```python
dept_attrition['Yes'].plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Attrition Rate by Department')
plt.ylabel('Attrition %')
plt.show()
```

**Attrition By Job Role**

```python
job_attrition = pd.crosstab(
    df['JobRole'],
    df['Attrition'],
    normalize='index'
)*100

print(job_attrition)
```

```python
job_attrition['Yes'].sort_values().plot(
    kind='barh',
    figsize=(10,6)
)

plt.title('Attrition Rate by Job Role')
plt.xlabel('Attrition %')
plt.show()
```

**Attrition By Income Group**

```python
income_attrition = pd.crosstab(
    df['Income Group'],
    df['Attrition'],
    normalize='index'
)*100

print(income_attrition)
```

```python
income_attrition['Yes'].plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Attrition by Income Group')
plt.ylabel('Attrition %')
plt.show()
```

**Attrition By Experience Group**

```python
experience_attrition = pd.crosstab(
    df['Experience Group'],
    df['Attrition'],
    normalize='index'
)*100

print(experience_attrition)
```

```python
experience_attrition['Yes'].plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Attrition by Experience Group')
plt.ylabel('Attrition %')
plt.show()
```

**Work Life Balance vs Attrition**

```python
sns.countplot(
    x='WorkLifeBalance',
    hue='Attrition',
    data=df
)

plt.title('Work Life Balance vs Attrition')
plt.show()
```

**Years Since Last Promotion vs Attrition**

```python
sns.boxplot(
    x='Attrition',
    y='YearsSinceLastPromotion',
    data=df
)

plt.title('Promotion Delay vs Attrition')
plt.show()
```

# Key Insights

1. Overall attrition rate is 16%.

2. Employees working overtime show a higher likelihood of leaving the company.

3. Lower income groups experience higher attrition.

4. Employees with lower job satisfaction tend to leave more frequently.

5. Some job roles have significantly higher attrition rates than others.

6. Poor work-life balance is associated with increased employee turnover.

7. Delayed promotions may contribute to employee exits.

8. Age and Monthly Income show a moderate positive relationship.#
